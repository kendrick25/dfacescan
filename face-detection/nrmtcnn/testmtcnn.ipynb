```python
from mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
```

```python
# Inicializar el detector MTCNN
detector = MTCNN()

# Cargar la imagen de entrada y convertirla a escala de grises (opcional, no se usa en MTCNN)
image = cv2.imread('../sampledata/others/img-01.png')
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir BGR a RGB para matplotlib

plt.figure(figsize=(20,20))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen Original')
plt.axis('on')

# Detectar los rostros usando MTCNN
faces = detector.detect_faces(image)

# Dibujar una caja delimitadora alrededor de cada rostro detectado
for face in faces:
    x, y, w, h = face['box']
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Convertir la imagen con las cajas delimitadoras a RGB para mostrarla con matplotlib
img_rgb_with_boxes = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Mostrar la imagen con los rostros detectados
plt.figure(figsize=(25,25))
plt.subplot(1, 2, 2)
plt.imshow(img_rgb_with_boxes)
plt.title('Rostros Detectados')
plt.axis('on')
plt.show()
```
