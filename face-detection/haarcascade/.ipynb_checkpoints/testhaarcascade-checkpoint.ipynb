```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#cargar los clasificadores pre-entrenados
face_cascade = cv2.CascadeClassifier('../libs/opencv-data/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('../libs/opencv-data/haarcascades/haarcascade_eye.xml')

#leer la imagen de un archivo
img = cv2.imread('../sampledata/crowds/group-people-9.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convertir BGR a RGB para matplotlib

#mostrar la imagen original
plt.figure(figsize=(24,24))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('')
plt.axis('on')
```

```python
# Convertir a escala de grises
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Detectar rostros en la imagen
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

for (x, y, w, h) in faces:
    # Dibujar un rectángulo alrededor del rostro
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 255, 0), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]

    # Detectar ojos dentro del rostro
    eyes = eye_cascade.detectMultiScale(roi_gray)

    # Dibujar un rectángulo alrededor de cada ojo
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 127, 255), 2)

# Convertir la imagen con detección de BGR a RGB para matplotlib
img_rgb_detected = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(24,24)) 
plt.subplot(1, 2, 2)
plt.imshow(img_rgb_detected)
plt.show()
```
